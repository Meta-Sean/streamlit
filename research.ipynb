```python
import pandas as pd
def load_data():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    html = pd.read_html(url, header = 0)
    df = html[0]
    return df
```

```python
df = load_data()
df
```

```python
sector = df.groupby('GICS Sector')
```

```python
import yfinance as yf
```

```python
df_selected_sector = sector.get_group('Health Care')
```

```python
data = yf.download(
        tickers = list(df_selected_sector[:10].Symbol),
        period = "ytd",
        interval = "1d",
        group_by = 'ticker',
        auto_adjust = True,
        prepost = True,
        threads = True,
        proxy = None
    )
```

```python
df = pd.DataFrame(data['ABT'].Close)
df['Date'] = df.index
df
```
